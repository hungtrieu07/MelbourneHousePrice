```python
mod_dt = dt.dropna( axis=0, how='any', subset=['Price'])
mod_dt
```

```python
mod_dt.isnull().sum()
```

```python
mod_dt.drop(['Address'], axis=1, inplace=True)
mod_dt
```

```python
for column in ['Rooms', 'Price', 'Distance', 'Postcode', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'YearBuilt','Lattitude', 'Longtitude', 'Propertycount' ]:
    mod_dt[column].fillna(mod_dt[column].median(), inplace=True)
```

```python
for column in ['Suburb', 'Type', 'Method', 'SellerG', 'CouncilArea', 'Regionname', 'Date']:
    mod_dt[column].fillna(mod_dt[column].mode()[0], inplace=True)
```
